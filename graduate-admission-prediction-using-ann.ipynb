```python
df.drop(columns=['Serial No.'],inplace=True)
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
try:
    import tensorflow
    from tensorflow import keras
    from keras import Sequential
    from keras.layers import Dense
    tf_available = True
except ImportError as e:
    print(f"TensorFlow import failed: {e}")
    print("Will load pre-trained model instead")
    tf_available = False
```

```python
if tf_available:
    model = Sequential()
    
    model.add(Dense(7,activation='relu',input_dim=7))
    model.add(Dense(7,activation='relu'))
    model.add(Dense(1,activation='linear'))
else:
    # Load pre-trained model
    try:
        import h5py
        model = keras.models.load_model("model.h5")
        print("Loaded pre-trained model from model.h5")
    except:
        print("Could not load pre-trained model. TensorFlow must be properly installed.")
```

```python
if tf_available:
    model.summary()
else:
    print("Model loaded from file. Summary not available without TensorFlow.")
```

```python
if tf_available:
    model.compile(loss='mean_squared_error', optimizer ='Adam')
else:
    print("Model compiled already. Skipping compile step.")
```

```python
if tf_available:
    history= model.fit(X_train_scaled,y_train,epochs=100,validation_split=0.2)
else:
    print("Model training skipped. Using pre-trained model.")
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
model.save("model.keras")
```
